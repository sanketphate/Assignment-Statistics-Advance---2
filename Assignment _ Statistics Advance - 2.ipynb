1. What is a random variable in probability theory?

A random variable is a numerical value determined by the outcome of a random experiment. It maps outcomes to numbers.


2. What are the types of random variables?


Discrete Random Variable: Takes specific, separate values (e.g., number of heads in 3 coin tosses).
Continuous Random Variable: Can take any value in a range (e.g., height, temperature).



3. What is the difference between discrete and continuous distributions?

Discrete distribution: Deals with countable outcomes.
Continuous distribution: Deals with uncountable, measurable outcomes.


4. What are probability distribution functions (PDF)?

For continuous random variables, the Probability Density Function (PDF) describes the likelihood of the variable taking a value within an interval. The area under the curve gives probability.


5. How do cumulative distribution functions (CDF) differ from PDFs?

CDF gives the probability that a variable is less than or equal to a value.
PDF shows the probability density at a specific value (for continuous) or probability (for discrete).



6. What is a discrete uniform distribution?

A distribution where all n outcomes have equal probability.
Example: Rolling a fair die → P(1) = P(2) = … = P(6) = 1/6.



7. What are the key properties of a Bernoulli distribution?

Only two outcomes: Success (1) or Failure (0)
Probability of success = p, failure = 1 − p
Mean = p, Variance = p(1 − p)



8. What is the binomial distribution, and how is it used in probability?

Models the number of successes in n independent Bernoulli trials with probability p.
Used in repeated yes/no experiments (e.g., 10 coin tosses).



9. What is the Poisson distribution and where is it applied?

Models the number of events occurring in a fixed interval of time or space when events occur independently at a constant average rate λ.
Used in call centers, traffic flow, etc.



10. What is a continuous uniform distribution?

Every value in a given interval has the same probability density.
Example: Selecting a random number between 0 and 1.



11. What are the characteristics of a normal distribution?

Bell-shaped, symmetric about the mean
Mean = Median = Mode
Fully defined by mean (μ) and standard deviation (σ)



12. What is the standard normal distribution, and why is it important?

A normal distribution with mean = 0, standard deviation = 1.
Used for calculating Z-scores and making statistical inferences.



13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The CLT states that the sampling distribution of the sample mean approaches a normal distribution as sample size n becomes large, regardless of the original distribution.
It allows us to use normal-based methods even for non-normal data.



14. How does the Central Limit Theorem relate to the normal distribution?

It explains why the normal distribution appears frequently in statistics: averages of large samples from any distribution tend to be normally distributed.



15. What is the application of Z statistics in hypothesis testing?

Z-statistics help compare a sample mean to a population mean under the assumption of known variance.
Used to calculate p-values and test significance.

17. What are point estimates and interval estimates in statistics?

Point estimate: A single value used to estimate a population parameter (e.g., sample mean).
Interval estimate: A range (like a confidence interval) that likely contains the true parameter.



18. What is the significance of confidence intervals in statistical analysis?

They provide a range of values within which we expect the true population parameter to lie, with a certain level of confidence (e.g., 95%).



19. What is the relationship between a Z-score and a confidence interval?

Z-scores determine the critical values used to construct confidence intervals.
E.g., for 95% CI → Z ≈ ±1.96.



20. How are Z-scores used to compare different distributions?

Z-scores standardize values across distributions, allowing comparison even when means and standard deviations differ.



21. What are the assumptions for applying the Central Limit Theorem?

Sample size is sufficiently large (typically n ≥ 30).
Samples are independent.
Population has finite mean and variance.

23. How does a probability distribution relate to the expected outcome of a random variable?

The probability distribution tells us how likely each outcome is, and the expected value summarizes the average outcome over many trials.
